```python
uk_2016 = pd.read_csv('../usedata/2016/UK_2016.csv')
data_trace = dict(
    type='sankey',
     domain = dict(
       x =  [0,1],
       y =  [0,1]
     ),
    orientation = "h",
    valueformat = ".0f",
    node = dict(
      pad = 10,
      thickness = 20,
      line = dict(
        color = "aliceblue",
        width = 0.3
      ),
      label =  uk_2016['Node, Label'].dropna(axis=0, how='any'),
      color = uk_2016['Color']
    ),
    link = dict(
      source = uk_2016['Source'].dropna(axis=0, how='any'),
      target = uk_2016['Target'].dropna(axis=0, how='any'),
      value = uk_2016['Value'].dropna(axis=0, how='any'),
        color=uk_2016['Link color'].dropna(axis=0,how='any'),
  )
)


fig = dict(data=[data_trace], layout=layout)
pio.show(fig)
```
